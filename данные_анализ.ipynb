```python
def remove_outliers(group):
    Q1 = group['За метр'].quantile(0.25)
    Q3 = group['За метр'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group['За метр'] >= lower_bound) & 
                (group['За метр'] <= upper_bound)]

sell_clean = sell.groupby('Город').apply(remove_outliers).reset_index(drop=True)
sell_clean.shape
```

```python
def annuity_payment(principal, annual_rate, years):
    monthly_rate = annual_rate / 12 / 100
    months = years * 12
    payment = principal * monthly_rate * (1 + monthly_rate)**months / ((1 + monthly_rate)**months - 1)
    return payment
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = sell_complete

rates = [6, 25]
loan_terms = [5, 10, 15, 20]
down_payment_ratio = 0.3

results = []
for city, price, rent, yield_ in zip(df['Город'], df['Цена'], df['Аренда'], df['Доходность, %']):
    loan_amount = price * (1 - down_payment_ratio)
    for rate in rates:
        for term in loan_terms:
            monthly_payment = annuity_payment(loan_amount, rate, term)
            coverage = "Да" if rent >= monthly_payment else "Нет"
            results.append({
                'Город': city,
                'Ставка, %': rate,
                'Срок, лет': term,
                'Платеж, руб': round(monthly_payment),
                'Аренда, руб': rent,
                'Покрытие': coverage
            })

results_df = pd.DataFrame(results)
print(results_df)

plt.figure(figsize=(12, 6))
for rate in rates:
    subset = results_df[results_df['Ставка, %'] == rate]
    plt.plot(subset['Срок, лет'], subset['Платеж, руб'], marker='o', label=f'Ставка {rate}%')

plt.axhline(y=df['Аренда'].max(), color='r', linestyle='--', label='Макс. аренда')
plt.xlabel('Срок ипотеки (лет)')
plt.ylabel('Ежемесячный платеж (руб)')
plt.title('Зависимость платежа от срока и ставки')
plt.legend()
plt.grid()
plt.show()
```
